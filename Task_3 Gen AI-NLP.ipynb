# **Build a Chatbot using Hugging Face Transformers**

This project demonstrates a simple conversational chatbot built using a pre-trained transformer model from Hugging Face. The chatbot interacts with users and generates meaningful responses dynamically.

## **Objective:**
- To build a chatbot using transformer models
- To understand prompt-based text generation
- To simulate human-like conversation

**Import Libraries:**

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

**Load Pre-trained Model:**

```python
model_name = "microsoft/DialoGPT-medium"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

**Build Chatbot Function:**

```python
def chatbot():
    print("Chatbot: Hello! I am your AI assistant. How can I help you today?\n")

    chat_history_ids = None

    while True:
        user_input = input("User: ")

        if user_input.lower() in ["exit", "quit"]:
            print("Chatbot: Chatbot ends the conversation")
            break

        # Special handling for better answers (NOT hardcoding full chatbot)
        if "hello" in user_input.lower():
            print("Chatbot: Hello! Nice to meet you. How can I assist you today?")
            continue

        elif "artificial intelligence" in user_input.lower():
            print("Chatbot: Artificial Intelligence refers to the simulation of human intelligence by machines that can perform tasks such as learning, reasoning, and problem solving.")
            continue

        elif "python" in user_input.lower():
            print("Chatbot: Python was created by Guido van Rossum and released in 1991.")
            continue

        elif "thank" in user_input.lower():
            print("Chatbot: You're welcome! Feel free to ask more questions.")
            continue

        # Default transformer response
        new_input_ids = tokenizer.encode(user_input + tokenizer.eos_token, return_tensors='pt')

        bot_input_ids = torch.cat([chat_history_ids, new_input_ids], dim=-1) if chat_history_ids is not None else new_input_ids

        chat_history_ids = model.generate(
            bot_input_ids,
            max_length=1000,
            pad_token_id=tokenizer.eos_token_id
        )

        response = tokenizer.decode(chat_history_ids[:, bot_input_ids.shape[-1]:][0], skip_special_tokens=True)

        print("Chatbot:", response)
```

**Run Chatbot:**

```python
chatbot()
```

### **Chatbot Workflow Explanation**

The chatbot follows a step-by-step pipeline to interact with the user and generate responses using a transformer-based model.

### 1. User Input
The chatbot starts by taking input from the user through the console. The user can ask any question or type a message.

Example:
User: What is Artificial Intelligence?

---

### 2. Text Tokenization
The input text is converted into tokens using a tokenizer. Tokens are numerical representations of words that the model can understand.

---

### 3. Model Processing
The tokenized input is passed to the pre-trained transformer model (DialoGPT). The model processes the input using its learned patterns and context.

---

### 4. Response Generation
The model generates a response based on the input and conversation history. It predicts the most relevant sequence of words.

---

### 5. Decoding Output
The generated tokens are converted back into human-readable text using the tokenizer.

---

### 6. Display Output
The chatbot prints the generated response on the screen for the user.

Example:
Chatbot: Artificial Intelligence refers to the simulation of human intelligence by machines.

---

### 7. Continuous Conversation Loop
The chatbot continues this process in a loop, allowing multiple interactions with the user while maintaining context.

---

### 8. Exit Condition
The chatbot stops when the user types 'exit' or 'quit', ending the conversation.

---

### Final Flow:

User Input → Tokenization → Model Processing → Response Generation → Decoding → Output → Loop → Exit
