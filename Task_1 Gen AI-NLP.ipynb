### **1: Conceptual Understanding**

**What is the difference between "Love" and "love" in NLP?**

In NLP, "Love" and "love" are considered different tokens because NLP is case-sensitive by default. Converting text to lowercase helps treat them as the same word.

**What happens if stopwords are not removed?**

The dataset becomes noisy and large. It may reduce model efficiency and accuracy because common words like "is", "the", "and" do not carry much meaning.

**Provide two real-world scenarios where removing stopwords can be harmful.**

- Sentiment analysis: "I am not happy" → removing "not" changes meaning.
- Question answering: "What is your name?" → removing "what" affects intent.

**Explain the difference between stemming and lemmatization.**

Stemming cuts words to root form (running → runn), often not meaningful.
Lemmatization converts words to proper root form (running → run), using vocabulary and grammar.

### **2: Build Advanced Preprocessing Function**

```python
import re

def preprocess_text(text):

    if not text or not isinstance(text, str):
        return [], ""

    text = re.sub(r"http\S+|www\S+|@\S+", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"(.)\1{2,}", r"\1", text)
    text = text.encode("ascii", "ignore").decode()
    text = text.lower()
    text = re.sub(r"[^a-z\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()

    tokens = text.split()
    tokens = [w for w in tokens if len(w) > 2 or w in ["no", "not"]]

    clean_sentence = " ".join(tokens)

    return tokens, clean_sentence


# 👇 IMPORTANT: ye function ke bahar hona chahiye
text = "I absolutely looooved this product 100%"

tokens, clean = preprocess_text(text)

print("Tokens:", tokens)
print("Clean:", clean)
```

### **3: Stress Testing**



```python
sentences = [
"Get 100% FREE access now!!!",
"I absolutely looooved this product 😍😍",
"Worst service ever... 0/10",
"Call me at 9876543210",
"This is THE best course!!!",
"Visit https://openai.com now!",
"Nooooo this is baaad!!!",
"OK OK OK I got it",
"Win $$$ now!!! Limited offer!!!",
"I am not happy with this"
]

results = []

for s in sentences:
    tokens, clean = preprocess_text(s)
    results.append((s, tokens, clean))

for r in results:
    print("Original:", r[0])
    print("Tokens:", r[1])
    print("Clean:", r[2])
    print()
```

### **4: Token Analytics**


```python
import numpy as np

for s, tokens, clean in results:
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = np.mean([len(t) for t in tokens]) if tokens else 0

    print("Sentence:", s)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Avg Token Length:", round(avg_len,2))
    print()
```

### **5: Frequency Analysis**

```python
from collections import Counter

all_tokens = []

for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 frequent:", counter.most_common(10))
print("Least frequent:", counter.most_common()[-5:])
```

### **6: Build Full Pipeline**


```python
def full_pipeline(text_list):

    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)   # ✅ FIXED

        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
sentences = [
"I absolutely looooved this product 😍😍",
"Visit https://openai.com now!",
"Nooooo this is baaad!!!"
]

result = full_pipeline(sentences)

print("\n--- FINAL PIPELINE OUTPUT ---\n")
print("All Tokens:")
print(result["tokens"])

print("\nClean Sentences:")
for i, sent in enumerate(result["clean_sentences"], 1):
    print(f"{i}. {sent}")
```

### **7: Error Handling**


```python
def preprocess_text(text):

    #  Handle None or invalid input FIRST
    if not text or not isinstance(text, str):
        return [], ""

    # Now safe to process
    text = text.lower()

    import re
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1+', r'\1', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    words = text.split()
    words = [w for w in words if len(w) > 2 or w in ["no", "not"]]

    cleaned = " ".join(words)
    tokens = cleaned.split() if cleaned else []

    return tokens, cleaned
def test_error_handling():

    test_cases = {
        "Empty String": "",
        "Only Emojis": "😍😍😍😍",
        "Only Numbers": "1234567890",
        "Spaces Only": "     ",
        "None Value": None
    }

    print("\n--- Task 7: Error Handling Results ---\n")

    all_passed = True  # flag to track overall success

    for name, input_text in test_cases.items():

        # Run preprocessing
        tokens, cleaned = preprocess_text(input_text)

        # Check if handled correctly (expect empty output)
        if tokens == [] and cleaned == "":
            status = "Handled"
        else:
            status = "Not Handled"
            all_passed = False

        # Print formatted output
        print(f"Test Case: {name}")
        print(f"  Input:  '{input_text}'")
        print(f"  Output: '{cleaned}'")
        print(f"  Status: {status}")
        print("-" * 40)

    # Final result
    if all_passed:
        print("\n All Edge Cases Passed! The Preprocessing Engine is robust.")
    else:
        print("\n Some test cases failed. Please check the pipeline.")

test_error_handling()
```
